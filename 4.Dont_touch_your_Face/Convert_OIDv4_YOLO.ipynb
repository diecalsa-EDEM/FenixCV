```python
import os
import cv2
from PIL import Image
import matplotlib.pyplot as plt
import time
%matplotlib inline
```

```python
def convert(size, box):
    dw = 1.0/size[0]
    dh = 1.0/size[1]
    x = (box[0] + box[2])/2
    y = (box[1] + box[3])/2
    w = box[2] - box[0]
    h = box[3] - box[1]
    
    x = x*dw
    y = y*dh
    w = w*dw
    h = h*dh
    
    return (x,y,w,h)
```

```python
start = time.time()
txt_files = []
os.chdir(LABEL_PATH)
# List labels in path
for filename in os.listdir(os.getcwd()):
    fname = filename.split('.')[0]
    #print(fname)
    if filename.endswith(".txt"):
        with open(filename,'r') as txt:
            text = txt.readlines()
        
            img_path = IMAGES_PATH + fname + '.jpg'
            # Get image size
            im=Image.open(img_path)
            
            w= int(im.size[0])
            h= int(im.size[1])
            
            print(w,h)
            for line in text:
                line_split = line.split(' ')
                xmin = line_split[len(line_split)-4]
                xmax = line_split[len(line_split)-3]
                ymin = line_split[len(line_split)-2]
                ymax = line_split[len(line_split)-1]
                
                b = (float(xmin), float(xmax), float(ymin), float(ymax))
                print(b)
                bb = convert((w,h), b)
                print(bb)
                write_txt = NEW_LABEL_PATH + fname + '.txt'
                with open(write_txt,'a') as wtxt:

                    wtxt.write('0 ' + str(bb[0]) + ' ' + str(bb[1]) + ' ' + str(bb[2]) + ' ' + str(bb[3]) + '\n')
os.chdir('/Users/Diego/Desktop/')
end = time.time()
print(end-start)
```
